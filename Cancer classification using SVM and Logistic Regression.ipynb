```python
def detect_outliers(cancer_data):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = np.percentile(cancer_data, 25)
    Q3 = np.percentile(cancer_data, 75)
    
    # Calculate the IQR
    IQR = Q3 - Q1
    
    # Define outlier bounds
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    
    # Identify outliers
    outliers = cancer_data[(cancer_data < lower_bound) | (cancer_data > upper_bound)]
    return outliers
```

```python
numerical_columns = cancer_data.select_dtypes(include='number').columns
for col in numerical_columns:
    outliers = detect_outliers(cancer_data[col])
    outlier_counts = len(outliers)
    print(f"No. of outliers in {col}: {outlier_counts}")
```
